```python
# Find the frequency of words in a webpage. Use urllib and BeautifulSoup to extract text from webpage.
```

```python
# Load modules

from bs4 import BeautifulSoup
import urllib.request
import nltk
```

```python
# Load data

response = urllib.request.urlopen('http://php.net/')
html = response.read()
```

```python
# Use BeautifulSoup to clean the grabbed text

soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)
```

```python
# Convert the text into tokens by splitting the text

tokens = [t for t in text.split()] 
```

```python
# Make a copy of the list

clean_tokens = tokens[:] 
```

```python
# Get English stop words

from nltk.corpus import stopwords
sr = stopwords.words('english') 
```

```python
# Iterate over the tokens and remove the stop words

for token in tokens: 
    if token in stopwords.words('english'): 
        clean_tokens.remove(token) 
```

```python
# Calculate the frequency distribution of those tokens using Python NLTK

freq = nltk.FreqDist(clean_tokens)
```

```python
# Print the frequency of words in the webpage

for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```
